## Задание 1
**Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.**

```python
import os, re

def lemmatize_into_file(input_file, output_file): #создаем новую функцию
    with open(input_file, 'r', encoding='utf-8') as f: #открываем и читаем наши анекдоты
        text = f.read()
   
    from pymystem3 import Mystem #лемматизируем
    m = Mystem()
    lemmas = m.lemmatize(text)
    lemmas_text =''.join(lemmas) #превращаем обратно в текст
   
    #записываем в новый файл
   
    with open(output_file, "w") as f:
         f.write(lemmas_text)
        
```

```python
lemmatize_into_file(input("Введите путь к начальному файлу\n"), input("Введите путь к конечному файлу\n")) 
#вызываем функцию
```

## Задание 2
**Очистите лемматизированный текст от стоп-слов и посчитайте ipm для оставшихся. Выведите 20 самых частотных по этому параметру слов.**

```python
words_text = [w.strip("!\"#$%&'()*+,./:;<=>?@[\]^_`{|}~„“«»†*—/\-—") for w in lemmas_text.split()] #чистим от пунктуации
words_text = [w for w in words_text if len(w) != 0]

with open("rus_stopwords.txt", 'r', encoding='utf-8') as f: #открываем файл со стоп-словами
    stopwords = f.read()
stopnew = stopwords.split("\n") #провели раздел по строке, чтобы слова шли друг за другом
new = [w for w in words_text if w not in stopnew] #тут ввели ограничение на то, чтобы в списке были слова не из стопслов
from collections import Counter #импортируем каунтер
D = Counter(new)
ipm_new = {} #превращаем в словарь
for k, v in D.items():
    ipm_new[k] = (v/len(new))*1000000 #считаем ipm
    
sort_ipm = sorted(ipm_new.items(), key=lambda  x: x[1], reverse=True) #выводим частотные слова
sort_ipm[:20]
```
