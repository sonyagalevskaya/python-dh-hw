## Задание 1

Дан текст (можно взять любой, необязательно этот):

## варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг
Необходимо выделить все биграммы в этом тексте и сохранить их список в какую-либо переменную. Биграммы -- это пары стоящих рядом слов, например, "варкалось хливкие", "хливкие шорьки" и т.д.

```python
text = '''варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как  мюмзики в мове о бо йся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг'''

text = text.split() 
bigrams = [] ##составляем список хранения биграмм

for i in range(len(text) - 1): #создаем биграммы для всех, кроме последнего слова
    bigram = (text[i], text[i + 1]) ##составляем биграммы из предыдущего слова и следующего
    bigrams.append(bigram) ##биграмма добавляется в список

for i in bigrams:
    print(i)
```

Пройтись циклом по рабочей директории, читая содержимое каждого файла и записывая его в новый файл poems.txt По завершении цикла должен получиться файл со всеми стихотворениями!

```python
import os
directory_path = "/Users/clumsysonya/Projects/python-for-dh/Classes/7/poems"
output_filepath = "/Users/clumsysonya/Projects/python-for-dh/Classes/7/poems.txt"

files_paths = []
#делаем цикл, чтобы побродить по директории и записывая в файл 
print("-- Файлы:")
for root, dirs, files in os.walk(directory_path):  
    for filename in files:
        print(filename)
        files_paths.append(os.path.join(root, filename))
print(f"-- Количество файлов: {len(files_paths)}")
print("-- Папки:")
for root, dirs, files in os.walk(directory_path):  
    for directory in dirs:
        print(directory)

open(output_filepath, 'w').close 

for i in files_paths:
    with open(i) as file_read:
        with open("poems.txt", "a") as file_append:
            for line in file_read:
                file_append.write(line) 
```

## Задание 3

Удивительная кошка

Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!


```python
#Скопировать текст, сохранить его в переменную.
#Узнать длину текста в символах.
text = '''
Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!'''
print('Количество символов: ', len(text))
```

```python
#Очистить текст от пунктуации, используя list compehensions. На выходе должен получиться список слов.
words = [w.strip("!?.,-") for w in text.split()]
print (words)
```

```python
#Объединить слова из получившегося списка в строку через пробел.
text_as_string = ' '.join(words)
print(text_as_string)
```

```python
#С помощью list comprehensions проверить, написано ли слово с большой буквы, 
#и если да, то привести его к нижнему регистру. На выходе должен получиться список из всех слов стихотворения,
#написанных с маленькой буквы (не только те, которые изначально были с маленькой буквы, а все).
words_lowercase = [i if not any(c.isupper() for c in i) else i.lower() for i in words]
print(words_lowercase)
```

```python
#Узнать длину текста в словах.
print('Количество слов:', len(words_lowercase))
```

```python
#Узнать количество уникальных слов.
setwords = set(words_lowercase) #мы сделали массив, чтобы не повторялись элементы
print('Количество уникальных слов :', len(setwords))
```

```python
#Распечатать последние 10 слов.
print(words_lowercase[-10:])
```

```python
#Соединить полученный в пункте 3 список в строку. Слова в строке должны быть написаны через знак переноса строки.
words_in_string_with_slash_n = ""
for i in words:
    words_in_string_with_slash_n += i + "\n"
print(words_in_string_with_slash_n)
```

```python
#Распечатать с 20 по 30 символ этой строки (включительно).
print(words_in_string_with_slash_n[19:30])
```

```python
#Разбить текст на биграммы, записать их в переменную и посчитать частотность каждой биграммы. 
#Биграммы должны быть в виде списка кортежей.
bigrams = [] #список для хранения биграмм
for i in range(len(words) - 1): #составляем цикл для всех слов, кроме последнего
    bigram = (words[i], words[i + 1]) #тут из предыдущего и следующего создаются биграммы
    bigrams.append(bigram) #биграмма добавляется в список

# считаем частотность каждой биграмммы
for f in bigrams:
    print('Данная биграмма ', f, 'встречается ', bigrams.count(f), "раз")
```
